```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/uniprot-human-reviewed.dataset-2', sep='\t')

df = df.set_index(['Entity 1', 'Entity 2'])
```

```python
df.index[100]
```

```python
import torch

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(df.values)

df.iloc[:] = scaler.transform(df)
```

```python
df.corr()
```

```python
correlation_mat = df.corr() ** 2

top_corr_features = correlation_mat.index

plt.figure(figsize=(20, 20))

sns.heatmap(correlation_mat, annot=True, cmap='RdYlGn')
```
